2)Write a program to generate employee ID-Card(image not required,id card show on console) 

```python

class Employee:
    __id=0
    __name=""
    __gender=""
    __city=""
    

# function to set data 
    def setData(self,id,name,gender,city):
        self.__id=id
        self.__name = name
        self.__gender = gender
        self.__city = city
        
	
	# function to get/print data
    def showData(self):
        print("****MNC ID CARD****")
        print("Id\t:",self.__id)
        print("Name\t:", self.__name)
        print("Gender\t:", self.__gender)
        print("City\t:", self.__city)
      

# main function definition
def main():
    #Employee Object
    emp=Employee()
    emp.setData(1,'Prathamesh','male','solapur')
    emp.showData()
	
if __name__=="__main__":
    main()
```
